```python
import pandas as pd
import numpy as np
from numpy import log, sqrt
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats 
```

```python
df = pd.read_feather('data/yellow_tripdata_01.feather')
df
```

```python
df["duration"] = df["duration"] / 60
```

```python
df = df.loc[(df["duration"] >= 0.5) & (df["total_amount"]>0)
            & (df["income"] > 0)]
```

```python
fig, ax = plt.subplots(1, 3) 
  
sns.distplot(df['duration'], hist = False, kde = True, 
            kde_kws = {'shade': True, 'linewidth': 2},  
            label = "duration", color ="blue", ax = ax[0]) 
  
sns.distplot(df['total_amount'], hist = False, kde = True, 
            kde_kws = {'shade': True, 'linewidth': 2},  
            label = "total_amount", color ="blue", ax = ax[1]) 

sns.distplot(df['income'], hist = False, kde = True, 
            kde_kws = {'shade': True, 'linewidth': 2},  
            label = "income", color ="blue", ax = ax[2]) 
  
plt.legend(loc = "upper right") 
fig.set_figheight(5) 
fig.set_figwidth(10) 
fig.set_figwidth(15) 
  
```

```python
df = df.loc[(df["duration"] <= 200) & (df["total_amount"]<=20000)
            & (df["income"] <=30000)]
```

```python
fig, ax = plt.subplots(1, 3) 
  
sns.distplot(df['duration'], hist = False, kde = True, 
            kde_kws = {'shade': True, 'linewidth': 2},  
            label = "duration", color ="blue", ax = ax[0]) 
  
sns.distplot(df['total_amount'], hist = False, kde = True, 
            kde_kws = {'shade': True, 'linewidth': 2},  
            label = "total_amount", color ="blue", ax = ax[1]) 

sns.distplot(df['income'], hist = False, kde = True, 
            kde_kws = {'shade': True, 'linewidth': 2},  
            label = "income", color ="blue", ax = ax[2]) 
  
plt.legend(loc = "upper right") 
fig.set_figheight(5) 
fig.set_figwidth(10) 
fig.set_figwidth(15) 
```
