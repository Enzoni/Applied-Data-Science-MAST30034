```python
import pandas as pd
import numpy as np
```

```python
file = ['data/yellow_tripdata_2016-01.csv',
        'data/yellow_tripdata_2016-02.csv',
        'data/yellow_tripdata_2016-03.csv',]
df1 = pd.read_csv(file[0],error_bad_lines=False)
df2 = pd.read_csv(file[1],error_bad_lines=False)
df3 = pd.read_csv(file[2],error_bad_lines=False)

df = df1.append(df1).append(df2)
df
```

```python
df.columns
```

```python
drop_col = ['VendorID','passenger_count','trip_distance', 'RatecodeID',
            'store_and_fwd_flag','payment_type', 'fare_amount','extra',
             'mta_tax', 'tip_amount','improvement_surcharge']

df.drop(drop_col, axis=1, inplace=True) 
```

```python
df.dropna(inplace=True)
df.reset_index(inplace=True)
df['start'] = pd.to_datetime(df['tpep_pickup_datetime'], format='%Y/%m/%d %H:%M', errors='coerce')
df['end'] = pd.to_datetime(df['tpep_dropoff_datetime'], format='%Y/%m/%d %H:%M', errors='coerce')
df['duration'] = (df['end'] - df['start']).dt.seconds.astype(int)
```

```python
# select useful information

df.to_feather('data/yellow_tripdata_01-03.csv.feather')
```
