```python
import pandas as pd
import numpy as np
from numpy import log, sqrt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_feather('data/yellow_tripdata_01.feather')
df
```

```python
sns.distplot(df['income'].apply(lambda x: log(x) if x else 0), bins=50)
plt.show()
```

```python
sns.distplot(df['income'].apply(lambda x: log(x) if x else 0), bins=30)
plt.show()
```

```python
df['duration'] = df['duration']/60
```

```python
df.describe().loc[['25%','75%']]
```

```python
def find_range(df):
    q1_q3 = df.describe().loc[['25%','75%']]
    step = q1_q3.loc['75%'] - q1_q3.loc['25%'] * 1.5
    print(step)
    q1_q3.loc['25%']  =  q1_q3.loc['25%'] - step
    q1_q3.loc['75%'] = q1_q3.loc['75%'] + step
    return q1_q3
```

```python
acc_range = find_range(df)
acc_range
```

```python
df_del_duration = df.loc[(df["duration"] >= 0) & (df["duration"] <= 70)]
```

```python
sns.distplot(df_del_duration['duration'],kde=False)

plt.show()
```

```python
df_del_duration = df.loc[(df["duration"] >= 0.5) & (df["duration"] <= 60)]
```

```python
df_total_amount = df_del_duration.loc[
    (df_del_duration["total_amount"] >= 0) &
    (df_del_duration["total_amount"] <= 80) &
    (df_del_duration["income"] >=0)]
```

```python
sns.distplot(df_total_amount['income'],kde=False)
plt.show()
```

```python
df_total_amount
```

```python
df_total_amount.drop(['tolls_amount','total_amount'],axis=1, inplace=True)
df_total_amount.reset_index(inplace=True)
```

```python
df_total_amount.to_feather('data/tripdata_with_weather.feather')
```
