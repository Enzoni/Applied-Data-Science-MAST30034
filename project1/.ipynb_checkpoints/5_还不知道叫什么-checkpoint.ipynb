```python
df = pd.read_feather('data/tripdata_with_weather.feather')
df.drop('index', axis=1, inplace=True) 
df
```

```python

df[['income', 'duration']].plot.scatter(x='income',y='duration',s=0.1)
plt.show()
```

```python
df['income/hour'] = df['in/du']  * 60*60
```

```python
def find_range(df):
    q1_q3 = df.describe().loc[['25%','75%']]
    step = q1_q3.loc['75%'] - q1_q3.loc['25%'] * 1.5
    print(step)
    q1_q3.loc['25%']  =  q1_q3.loc['25%'] - step
    q1_q3.loc['75%'] = q1_q3.loc['75%'] + step
    return q1_q3
acc_range = find_range(df)
acc_range
```

```python
df = df.loc[df['income/hour'] <= 250]
```

```python
sns.distplot(df['income/hour'],kde=False)

plt.show()
```
