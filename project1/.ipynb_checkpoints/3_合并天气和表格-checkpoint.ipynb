```python
import pandas as pd
import numpy as np
```

```python
weather = pd.read_csv('data/weather_preprocessed.csv', index_col= 'time')
weather
```

```python
car_record = pd.read_feather('data/yellow_tripdata_01-02.feather')
```

```python
car = car_record.iloc[:1000,:]
car
```

```python
car['start_hour'] =  car['start'].dt.hour
```

```python
car['start_date'] = car['start'].dt.strftime('%Y-%m-%d')
```

```python
car
```

```python
#car['weather'] = weather.loc[,car['start_hour']]
```

```python
def fill_weather(weather, date, time):
    return weather.loc[date][time]


car['weather'] = car[['start_date','start_hour']].apply(
    lambda x: fill_weather(weather,x.iloc[0],x.iloc[1]), axis=1)

car
```
